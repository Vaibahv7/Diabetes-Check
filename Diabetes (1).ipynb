## DATA Preprocessing And Feature Eng



```python
# Step 1
data.isnull().sum()
```

```python
##Step 2 Handling the corrupted data
# In 'Glucose','BloodPressure','SkinThickness','Insulin','BMI' certain

data.Glucose.replace(0,np.median(data.Glucose),inplace=True)

#dataframe.column.replace
```

```python
data.loc[data['BMI']==0]
```

```python
data.loc[data['Glucose']==0]
```

```python
data.BloodPressure.replace(0,np.median(data.BloodPressure),inplace=True)
data.loc[data['BloodPressure']==0]
```

```python
data.SkinThickness.replace(0,np.median(data.SkinThickness),inplace=True)
data.loc[data['SkinThickness']==0]
```

```python
data.Insulin.replace(0,np.median(data.Insulin),inplace=True)
data.loc[data['Insulin']==0]
```

```python
data.BMI.replace(0,np.median(data.BMI),inplace=True)
data.loc[data['BMI']==0]
```

```python
X=data.iloc[:,:-1]
X=X.iloc[:,1:]
X
```

```python
y=data.iloc[:,-1]
y
```

```python
# Step 1 creating independent and dependant variable

## X=final_df.iloc[:,:-1]


```

```python
#Step 2 Creating Training and testing data

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=45)
```

```python
y_test
```

```python
# Step 3 Model creation
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()    #CLF model name
clf.fit(X_train,y_train)
```

```python
# Step 4
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
y_pred_prob=clf.predict_proba(X_test)
```

```python
y_pred_prob
```

```python
from sklearn.metrics import accuracy_score   # Checking accuracy
```

```python
accuracy_score(y_test,y_pred)
```
